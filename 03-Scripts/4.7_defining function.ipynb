# Create price label def function with If Statements


```python
# Create price label with def function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying new price range flag with defined function price_label 
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Count the price_values
df['price_range'].value_counts(dropna = False)
```

```python
# Check max price to confirm
df['prices'].max()
```

# Create price label with if-statements & loc() function

```python
# defining high prices range with loc()
df_ords.loc[df_ords['prices'] > 15, 'price_range_loc'] = 'high-range product'
```

```python
# defining mid prices range with loc()
df_ords.loc[(df_ords['prices'] <= 15) & (df_ords['prices'] > 5), 'price_range_loc'] = 'mid-range product'
```

```python
# defining low prices range with loc()
df_ords.loc[df_ords['prices'] <= 5, 'price_range_loc'] = 'low-range product'
```

```python
# looking for counta per price range
df_ords['price_range_loc'].value_counts(dropna = False)
```

# Create flag: busiest day with Loop

```python
# Create new column in dataframe
df_ords['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Createing loop for busiest days
result = []

for value in df_ords["orders_day_of_week"]:

  if value == 0:
    result.append("busiest day")
  elif value == 4:
    result.append("least busy")
  else:
    result.append("regularly busy")
```

```python
# Check loop result
result
```

```python
# Add result to dataframe
df_ords['busiest_day'] = result
```

```python
# Check counts
df_ords['busiest_day'].value_counts(dropna = False)
```

# Create new flag busiest days (2 busiest days of the week)


```python
# Create new if-statement for-loop for 2 busiest days
result_2_days = []

for value in df_ords["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2_days.append("Busiest days")
  elif value == 4 or value == 3:
    result_2_days.append("Least busy days")
  else:
    result_2_days.append("Regularly busy")
```

```python
# add to dataframe
df_ords['busiest_days'] = result_2_days
```

```python
# checking
df_ords.head()
```

# Create flag: busiest_period_of_day

```python
# Checking counts of orders by hour of the day
df_ords['order_hour_of_day'].value_counts()
```

### NOTE:  To meet the client's request for segmentation, we can categorize the 24-hour day into three segments as outlined: High-order periods: 10, 11, 14, 15, 13, 12, 16, 9; Moderate-order periods: 17, 8, 18, 19, 20, 7, 21, 22; Low-order periods: 23, 6, 0, 1, 5, 2, 4, 3.

```python
# Create a loop if statement using the above breakdown
hour_of_day = []

for value in df_ords["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_day.append("Fewest orders")
  else:
    hour_of_day.append("Average orders")
```

```python
# Add result to dataframe
df_ords['busiest_period_of_day'] = hour_of_day
```

```python
# Checking
df_ords.head()
```

```python
# Frequency of busy hours
df_ords['busiest_period_of_day'].value_counts()
```
